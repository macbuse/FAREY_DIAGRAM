- d'Ocagne's Identity$F_m F_{n+1} - F_n F_{m+1} = (-1)^n F_{m-n}$
- $\Rightarrow$ Cassini's Identity$F_n^2 - F_{n-1}F_{n+1} = (-1)^{n-1}$

- Catalan's Identity$F_n^2 - F_{n-r}F_{n+r} = (-1)^{n-r}F_r^2$


After applying a suitable power of [[0,1],[1,1]]

### d'Ocagne's identity 

can assume 

$$\det \begin{pmatrix}
F_m & F_n \\
F_{m+1} & F_{n+1} 
\end{pmatrix}
= 
\det \begin{pmatrix}
0 & 1 \\
1 & 1 
\end{pmatrix}^{-m}
\det \begin{pmatrix}
0 & F_{n-m} \\
1 & F_{n-m+1} 
\end{pmatrix}
= (-1)^m
\det
\begin{pmatrix}
0 & F_{n-m} \\
1 & F_{n-m+1} 
\end{pmatrix}
= (-1)^m F_{n-m}
$$
