```python
import numpy as np
import matplotlib.pyplot as plt


def farey(n):
    F_n = np.array([[0,1], [1,1]],dtype=int)
    for k in range(n):
        tmp = np.zeros((2*F_n.shape[0]-1, 2),dtype=int)
        # copy F_n in 
        tmp[::2,] = F_n
        # add medians in
        tmp[1::2,] = F_n[:-1] + F_n[1:]
        F_n = tmp
    return F_n
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parametric equations for the Astroid
def astroid(t, a=1):
    x = a * np.cos(t)**3
    y = a * np.sin(t)**3
    return x, y

# Generate parameter values
t_values = np.linspace(0, 2 * np.pi, 400)  # t ranges from 0 to 2*pi
x_values, y_values = astroid(t_values)

# Plot the Astroid
#plt.figure(figsize=(6,6))
plt.plot(x_values, y_values, label="Astroid", color="blue")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.title("Astroid: $x = a \cos^3(t), y = a \sin^3(t)$")
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal', adjustable='box')
#plt.show()
plt.savefig("asteroid.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate parameter values
T = np.linspace(0, 2 * np.pi, 100)  # t ranges from 0 to 2*pi
X = np.cos(T)
Y = .5*np.sin(2*T)

# Plot the Figure-8 curve
#plt.figure(figsize=(6,6))
plt.plot(X,Y, label="Figure-8 Curve", color="blue")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.title("Figure-8 Curve: $x = \cos(t), y = 1/2 \cdot \sin(2t)$")
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal', adjustable='box')
#plt.show();
plt.savefig("fig8.png")
```
